```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,1,200)
```

```python
for i in range (1,20):
    plt.plot(t, np.sin(2*np.pi*t-t[10*i]), 'g-')
    plt.plot(t, np.cos(2*np.pi*t-t[10*i]), 'r-')
    plt.title("Exercise example")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")

plt.show()
```

```python
from scipy.stats import skewnorm, norm
from ipywidgets import interact
from ipywidgets import widgets

# Start by fattening the function distribution here
def plot_norm(skew_para=0, mean=0, sd=5):
    data_X = np.random.normal(mean, sd, 100000)
    plt.hist(data_X,bins=100)
    t = (data_X - mean) / sd
    plt.hist((2*norm.pdf(t)*norm.cdf(skew_para*t)),bins=100)
    #plt.hist(((data_X - mean)/sd),bins=100)
    #skew_pdf = 2 * norm.pdf(data_X) * norm.cdf(skew_para * data_X)
    #plt.hist(skew_pdf)
    plt.xlim((-15,15))
    plt.title("Skewnorm")

# Add a new parameter to control the fatness of the distribution side
interact(plot_norm, a=(0,5,1), mean=(-5,5,10), sd=(0,30,5))
```
